# Définition des constantes de simulation

```python
TSim = 2000
Ts = 1
N = int(TSim/Ts) + 1 

#MV

Kp_OMV_FOPDT = 0.5280686804191848
T_OMV_FOPDT = 160.98841281582534
theta_OMV_FOPDT = 35.21872295460488

Kp_OMV_SOPDT = 0.5279783321046589
T1_OMV_SOPDT = 156.72135296437907
#160.37049931289846
T2_OMV_SOPDT = 22.414015717250276
# 1.5102440400334622
theta_OMV_SOPDT = 15.732227706547503
#34.601306808660546

#DV 

Kp_ODV_FOPDT = 0.6153119961469684
T_ODV_FOPDT = 163.91357248141884
theta_ODV_FOPDT = 32.38845620532809

Kp_ODV_SOPDT =  0.611563683291308
T1_ODV_SOPDT = 155.0912717619001
T2_ODV_SOPDT = 25.53958809347822
theta_ODV_SOPDT = 11.559549786267498




#Operating points 
DV0 = 50 
MV0 = 50
PV0 = 77.3

# Set maximum and minimum MV values
MVmin = 0
MVmax = 100

# Coefficients
alpha = 1
gamma = 0.6
```

# Choix du scénario de simulation

```python


scenario = 8 

# 1: Boucle ouverte sans feedforward (OLP + no FF)
if scenario == 1:
    SPPath = {0: PV0}
    ManPath = {0: True, TSim: True}
    MVManPath = {0: MV0, TSim: MV0}
    DVPath = {0: DV0, 1300: DV0 + 10}
    FF = False
    ManFF = True

# 2: Boucle ouverte avec feedforward (OLP + FF)
elif scenario == 2:
    SPPath = {0: PV0}
    ManPath = {0: True, TSim: True}
    MVManPath = {0: MV0, TSim: MV0}
    DVPath = {0: DV0, 1300: DV0 + 10}
    FF = True
    ManFF = True

# 3: Boucle fermée sans feedforward (CLP + no FF)
elif scenario == 3:
    SPPath = {0: PV0 + 5, 700: PV0 - 5, TSim: PV0}
    ManPath = {0: True, 500: False, TSim: False}
    MVManPath = {0: MV0, TSim: MV0}
    DVPath = {0: DV0, 1300: DV0 + 10}
    FF = False
    ManFF = False

# 4: Boucle fermée avec feedforward (CLP + FF)
elif scenario == 4:
    SPPath = {0: PV0 + 5, 750: PV0 - 5, TSim: PV0}
    ManPath = {0: True, 500: False, TSim: False}
    MVManPath = {0: MV0 + 10, TSim: MV0 + 10}
    DVPath = {0: DV0, 1000: DV0 + 10}
    FF = True
    ManFF = False

# 5: Feedforward seul (pas de PID, mode manuel actif)
elif scenario == 5:
    SPPath = {0: PV0, 600: PV0 + 10}
    ManPath = {0: True, TSim: True}
    MVManPath = {0: MV0, TSim: MV0}
    DVPath = {0: DV0, 1000: DV0 + 10}
    FF = True
    ManFF = True

# 6:  Kc trop élevé
elif scenario == 6:
    SPPath = {0: PV0 + 10}
    ManPath = {0: True, 200: False, TSim: False}
    MVManPath = {0: MV0, TSim: MV0}
    DVPath = {0: DV0, 1300: DV0 + 10}
    FF = False
    ManFF = False
    

# 7: DV change brutalement (test de robustesse PID seul)
elif scenario == 7:
    SPPath = {0: PV0 + 5}
    ManPath = {0: True, 300: False, TSim: False}
    MVManPath = {0: MV0, TSim: MV0}
    DVPath = {0: DV0, 1000: DV0 + 20}  
    FF = False
    ManFF = False

# 8: Rampe de consigne (test de suivi lent)
elif scenario == 8:
    SPPath = {0: PV0, 600: PV0 + 10}  
    ManPath = {0: True, 300: False, TSim: False}
    MVManPath = {0: MV0, TSim: MV0}
    DVPath = {0: DV0, 1200: DV0 + 10}
    FF = True
    ManFF = False



   


```

```python
# # IMC tuning
Kc, Ti, Td = IMC_Tuning(Kp_OMV_SOPDT, T_OMV_FOPDT, theta_OMV_SOPDT, gamma, order=1)
print(f"Kc: {Kc}, Ti: {Ti}, Td: {Td}")
```

# Tracé des résultats de la simulation

```python


# Créer la figure et les sous-graphiques
fig, (ax1, ax2, ax3, ax4) = plt.subplots(4, 1, gridspec_kw={'height_ratios': [1, 10, 10, 1]})
fig.set_figwidth(24)
fig.set_figheight(12)

# Tracé du mode manuel
l1, = ax1.step([0, TSim], [0, 1], 'k-', label="Man", where='post')

# === TITRE SCÉNARIO ===
scenario_titles = {
    1: 'Scenario 1: Open-loop response without PID and without feedforward',
    2: 'Scenario 2: Open-loop response without PID but with feedforward',
    3: 'Scenario 3: Closed-loop PID control without feedforward',
    4: 'Scenario 4: Closed-loop PID control with feedforward',
    5: 'Scenario 5: Feedforward-only control (manual mode, no PID)',
    6: 'Scenario 6: Closed-loop with poorly tuned PID (high Kc)',
    7: 'Scenario 7: Closed-loop PID with strong disturbance (DV +20)',
    8: 'Scenario 8: Closed-loop with ramp setpoint and feedforward'
}

# Titre spécifique au scénario
ax1.set_title(scenario_titles.get(scenario, f'Scenario {scenario}'))

# Titre global de la figure
plt.suptitle(f"Simulation Test – Scenario {scenario}", fontsize=14, weight='bold')

# Légende pour l'axe ax1
ax1.legend(loc='best')


l2, = ax2.step([0,TSim],[0,100],'b-',linewidth=2,label="MV",where='post')
l3, = ax2.step([0,TSim],[0,100],'y--',linewidth=1,label="MVP",where='post')
l4, = ax2.step([0,TSim],[0,100],'c--',linewidth=1,label="MVI",where='post')
l5, = ax2.step([0,TSim],[0,100],'b--',linewidth=1,label="MVD",where='post')
l6, = ax2.step([0,TSim],[0,100],'r--',linewidth=1,label="MVFF",where='post')
ax2.set_ylabel("Value of MV [%]")
ax2.legend(loc='best')
ax2.grid()

l7, = ax3.step([0,TSim],[0,100],'k-',linewidth=2,label="SP",where='post')
l8, = ax3.step([0,TSim],[0,100],'g-',linewidth=1,label="PV",where='post')
ax3.set_ylabel("Value of PV [°C]")
ax3.legend(loc='best')
ax3.grid()

l9, = ax4.step([0,TSim],[0,100],'r-',linewidth=2,label="DV",where='post')
ax4.set_xlabel("Time [s]")
ax4.set_ylabel("Value of DV [%]")
ax4.legend(loc='best')

# Running experiment with chosen scenario

t = []

SP = []
PV = []
MAN = []
MV_MAN = []
DV = []
MVFF = []
MV = []
MVp = []
MVi = []
MVd = []
E = []
PV_p = []
PV_d = []

MVFF_Delay = []
MVFF_LL1 = []
MV_Delay_P = []
MV_FO_P = []
MV_Delay_D = []
MV_FO_D = []

lab = tclab.TCLab()

for i in range(0, N):
    t.append(i * Ts)
    
    # If first iteration, set last_time to current time
    if t[-1] == 0:
        last_time = time.time()
        
    SelectPath_RT(SPPath, t, SP)
    SelectPath_RT(ManPath, t, MAN)
    SelectPath_RT(MVManPath, t, MV_MAN)
    SelectPath_RT(DVPath, t, DV)
    
    # Process Value
    PV.append(lab.T1)
    
    # Apply DV
    lab.Q2(DV[-1])
    
    # FeedForward
    Delay_RT(DV - DV0*np.ones_like(DV), max(theta_ODV_SOPDT-theta_OMV_SOPDT, 0), Ts, MVFF_Delay)
    LL_RT(MVFF_Delay, -Kp_ODV_SOPDT/Kp_OMV_SOPDT, T1_OMV_SOPDT, T1_ODV_SOPDT, Ts, MVFF_LL1)
    if FF == True:
        LL_RT(MVFF_LL1, 1, T2_OMV_SOPDT, T2_ODV_SOPDT, Ts, MVFF)
    else:
        LL_RT(MVFF_LL1, 0, T2_OMV_SOPDT, T2_ODV_SOPDT, Ts, MVFF) # Set MVFF to 0 if FF is disabled
    
    # PID
    PID_RT(SP, PV, MAN, MV_MAN, MVFF, Kc, Ti, Td, alpha, Ts, MVmin, MVmax, MV, MVp, MVi, MVd, E, ManFF, PV0)
    
    # Apply MV
    lab.Q1(MV[-1])
    
    # Wait for sampling period Ts
    now = time.time()
    elapsed = now - last_time
    time.sleep(np.max([0, Ts - elapsed]))
    last_time = time.time()
    
    # Set Data
    l1.set_data(t,MAN)
    l2.set_data(t,MV)
    l3.set_data(t,MVp)
    l4.set_data(t,MVi)
    l5.set_data(t,MVd)
    l6.set_data(t,MVFF)
    l7.set_data(t,SP)
    l8.set_data(t,PV)
    l9.set_data(t,DV)
    
    clear_output(wait=True)

    ax1.set_xlim(0, t[-1]+1)
    ax2.set_xlim(0, t[-1]+1)
    ax3.set_xlim(0, t[-1]+1)
    ax4.set_xlim(0, t[-1]+1)
        
    display(fig)
    
lab.close() # Disconnect from TCLab

MVMinLimit = myRound(np.min((np.min(MV),np.min(MVp),np.min(MVi),np.min(MVd),np.min(MVFF))),5)
MVMaxLimit = myRound(np.max((np.max(MV),np.max(MVp),np.max(MVi),np.max(MVd),np.max(MVFF))),5)

ax1.set_ylim(-0.1,1.1)
ax2.set_ylim(MVMinLimit-10, MVMaxLimit+10)
ax3.set_ylim(myRound(np.min((np.min(PV),np.min(SP))),5)-5, myRound(np.max((np.max(PV),np.max(SP))),5)+5)
ax4.set_ylim(myRound(np.min(DV),5)-5, myRound(np.max(DV),5)+5)

# Save plot
now = datetime.now()
date_time = now.strftime("%Y-%m-%d-%Hh%M")

if not os.path.exists('Plots'):
    os.makedirs('Plots')
plt.savefig(f'Plots/Experiment_scenario_{scenario}_{date_time}.png',transparent=True)
plt.savefig(f'Plots/Experiment_scenario_{scenario}_{date_time}.pdf',transparent=True)

# Save data
t = np.array(t)
MAN = np.array(MAN)
MV = np.array(MV)
MVp = np.array(MVp)
MVi = np.array(MVi)
MVd = np.array(MVd)
MVFF = np.array(MVFF)
SP = np.array(SP)
PV = np.array(PV)
DV = np.array(DV)

data = np.vstack((t.T, MAN.T, MV.T, MVp.T, MVi.T, MVd.T, MVFF.T, SP.T, PV.T, DV.T))
data = data.T
if not os.path.exists('Data'):
    os.makedirs('Data')
nameFile = f'Data/Experiment_scenario_{scenario}_{date_time}.txt'
np.savetxt(nameFile,data,delimiter=',',header='t,MAN,MV,MVp,MVi,MVd,MVFF,SP,PV,DV',comments='')
```
